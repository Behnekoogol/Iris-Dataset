# Project: Iris Dataset

## Load Data 

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline 
```

```python
sns.set_theme(
    style = 'whitegrid',
    font_scale = 1.2
)
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

## Analyzing the Data 
> 5 comumns are available in this DataFrame

```python
df.columns.values
```

## Check Data Types

```python
df.dtypes
```

```python
df.species.value_counts()
```

```python
sns.countplot(
    data=df,
    y='species'
)
```

> converting 'species' into Categorcal Data Type

```python
df = df.astype(
    {
        'species': 'category'
    }
     
)
```

```python
df.dtypes
```

```python
df.describe()
```

## Visualizing Data

```python
sns.pairplot(
    data=df, 
    hue='species'
)
```

```python
g = sns.PairGrid(data=df, hue='species')
g.map_upper(sns.histplot)
g.map_diag(sns.kdeplot, fill=True)
g.map_lower(sns.scatterplot)
```

## Correlation 

```python
sns.heatmap(df.corr(), 
        cmap='coolwarm',  
        vmax=1, 
        vmin=-1,
        center=0,
        square=True,
        linewidths=.5,
        cbar_kws={"shrink": .5},
        annot=True,)
```
